```python
import pandas as pd
```

```python
dataset = pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
dataset.columns
dataset = pd.get_dummies(dataset , dtype = int,  drop_first=True)
```

```python
independence = dataset[['age','bmi', 'children','sex_male', 'smoker_yes']]
dependence = dataset[['charges']]
```

```python
dependence
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independence, dependence, test_size = 0.30, random_state = 0)
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel = "linear",C=3000)
regressor=regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
r_score
```

```python
import pickle 
filename = "final_modal_svm.sav"
```

```python
pickle.dump(regressor,open(filename,'wb'))
```

```python
loaded_model = pickle.load(open("final_modal_svm.sav",'rb'))
result = loaded_model.predict([[20 , 27.978 , 0 , 1 , 0]])
```
