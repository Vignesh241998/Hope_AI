```python
# It's a library that having the functions releated to the Database 
import pandas as pd
```

```python
# Dataset is read the csv file using pandas and assignes to the Dataset variable . 
dataset = pd.read_csv("Salary_Data.csv")

# Display the dataset 
dataset 
```

```python
# It is the Input and Output split logic . YearsExperience Column is split from the dataset and assigned to the independent variable
independent = dataset[["YearsExperience"]]

# Display the independent 
independent
```

```python
# It is the Input and Output split logic . Salary Column is split from the dataset and assigned to the dependent variable
dependent = dataset[["Salary"]]

# Display the dependent 
dependent
```

```python
# It is the Train & Test Set logic. 
#sklearn is the python library used for Machine Learning . model_selection is the package inside sklearn used to split data
#train_test_split is the function inside the model_selection for split train & Test data . 
from sklearn.model_selection import train_test_split

# The train_test_split function takes independent variables (X) and dependent variables (Y) along with parameters like test_size and random_state.
# It splits the data into training and testing sets and returns four outputs
# which are assigned to x_train, x_test, y_train, and y_test.
x_train,x_test,y_train,y_test = train_test_split(independent , dependent , test_size=0.30 , random_state=0)
```

```python
# This is a built-in algorithm for Linear Regression . 
from sklearn.linear_model import LinearRegression

#  Creating a model object and assignes to the regressor variable
regressor = LinearRegression()

# It is the modal training logic . fit() is the function to train the modal with the parameter actual input and output . 
regressor.fit(x_train,y_train)
```

```python
# After model created , take the weight(slope) and assigned in the weight variable 
weight = regressor.coef_

# Diplay the weight
weight
```

```python
# After model created , bias(Intercept) value taken and assigned in the bias variable
bias = regressor.intercept_

# display the bias 
bias
```

```python
# Evaluation Metrices to test the created modal is correctly working or not 
y_pred = regressor.predict(x_test)
```

```python
# It Calculate r2 score by comparing actual values (y_test) and predicted values (y_pred)
from sklearn.metrics import r2_score

# Pass the actual value and predicted value as parameter in r_score function
r_score = r2_score(y_test,y_pred)

# Display the r_score . If it is 1 means it is a good model else bad model.
r_score
```

```python
# Save the model logic . pickle is the library to save the python in file .  
import pickle 

# Create a variable to store the file name with .sav extension 
filename = "final_modal_linear.sav"
```

```python
# .dump() is the function to save the model . regressor is the trained model , filename is the created file name and wb is the write binary mode . 
pickle.dump(regressor,open(filename,'wb'))
```

```python
# Load the saved model from the file and assigned to loaded_model variable. 
# open() is the open the saved file , final_modal_linear.sav is file name and rb is read binary.

loaded_model = pickle.load(open("final_modal_linear.sav",'rb'))

# To test the created model , pass the value as input and assigned it in result variable
result = loaded_model.predict([[13]])
```
