```python
# Odd numbers in the list

numbers = [20,10,16,19,25,1,276,188]

for odd in numbers:
    if(odd % 2 == 1):
        print(odd ,"is Odd")
```

```python
# Even numbers in the list

numbers = [20,10,16,19,25,1,276,188]

for even in numbers:
    if(even % 2 == 0):
        print(even ,"is Even")
```
