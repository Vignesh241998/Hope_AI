```python
#8.Find the error
for i in range(1, 10, -1):
    print(i)

Issue : 
The step value is negative but the range is increasing. 

Explanation : 
A negative step means the loop should go backwards, but the range given is forward (1 to 10).
So the loop will not execute by this code . 

Correct code for backward loop : 

for i in range(10, 1, -1):
    print(i)
```

```python
#9. Find the error in the following Python code:
x = 10
if x = 5:
    print("x is 5")

Issue : 
Syntax Error 

Explanation : 
Because in if condition , there need to check with '==' but in this check with '=' .

Correct Code : 
x = 10
if x == 5:
    print("x is 5")
```
