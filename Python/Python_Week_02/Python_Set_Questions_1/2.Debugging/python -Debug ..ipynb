```python
#1.Find the error in the following Python code:

 for i in range(5)
    print(i)


Explanation : 
Issue: Missing colon : after the for loop.

Correct Code : 
for i in range(5):
    print(i)
```

```python
#2.What will be the output of this code?

a = 10
b = "5"
print(a + b)

Output : 
TypeError: unsupported operand type(s) for +: 'int' and 'str'

Explanation : 
Cannot add int and str data types in Python. Both values must be the same type.

Correct Ways of Code: 
1. a = 10
b = "5"
print(a + int(b))

2. a = 10
b = "5"
print(str(a) + b)
 
```

```python
#8.Find the error
for i in range(1, 10, -1):
    print(i)

Issue : 
The step value is negative but the range is increasing. 

Explanation : 
A negative step means the loop should go backwards, but the range given is forward (1 to 10).
So the loop will not execute by this code . 

Correct code for backward loop : 

for i in range(10, 1, -1):
    print(i)
```

```python
#9. Find the error in the following Python code:
x = 10
if x = 5:
    print("x is 5")

Issue : 
Syntax Error 

Explanation : 
Because in if condition , there need to check with '==' but in this check with '=' .

Correct Code : 
x = 10
if x == 5:
    print("x is 5")
```
